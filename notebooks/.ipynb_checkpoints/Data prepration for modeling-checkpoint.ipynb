```python
# handling missing values
train_df['sell_price'].fillna(0, inplace=True)
train_df['revenue'].fillna(0, inplace=True)

test_df['sell_price'].fillna(0, inplace=True)
test_df['revenue'].fillna(0, inplace=True)
```

# Feature engineering 

Encoding

```python
train_df.columns
```

```python
from sklearn.preprocessing import OrdinalEncoder

cat_col = ['item_id', 'dept_id', 'cat_id', 'store_id', 'state_id', 'count',
        'wm_yr_wk', 'event_name', 'event_type']
num_col = [e for e in train_df.columns if e not in cat_col + ['revenue','id']] 

 

# encode categorical features
for i, feature in enumerate(cat_col):
    train_df[feature] = OrdinalEncoder(dtype="int").fit_transform(train_df[[feature]])
```

```python
# encode categorical features
for i, feature in enumerate(cat_col):
    test_df[feature] = OrdinalEncoder(dtype="int").fit_transform(test_df[[feature]])
```

# Split dataset

```python
# indicating x and y columns 
y_train= train_df["revenue"]
x_train= train_df.drop(["revenue"],axis=1)

y_test=test_df["revenue"]
x_test=test_df.drop(["revenue"],axis=1)
```

```python
#Save the sets into the folder `data/processed`
# Solution
x_train.to_csv('../data/processed/x_train.csv', index=False)
x_test.to_csv('../data/processed/x_test.csv', index=False)
y_train.to_csv('../data/processed/y_train.csv', index=False)
y_test.to_csv('../data/processed/y_test.csv', index=False)
```
